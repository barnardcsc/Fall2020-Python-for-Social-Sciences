##Create a Column that Bins the Data
Create a new column that contains data that maps Freedom values to Low, Medium, and High.

````
#use the describe method on the Freedom column to find values for low, medium, and high
df_2015['Freedom'].describe()
````

####Create a new categorical variable using the 25% and 75% values from the describe method on Freedom to separate into 'low', 'medium', and 'high' brackets. 

####Add Columns

The Dataframe, *our_dataframe*, is just a variable.   To grab a column of a Dataframe we write:

```
our_dataframe['column name'] = values

```
To grab a create a bin in a column of a Dataframe we write:

```
our_dataframe.loc[ which values, 'column name'] = 'bin_name'
```

We can create a new column on the Dataframe variable by assiging a value to the Dataframe variable on a new column name:



````
df_2015.loc[df_2015['Freedom'] <= 0.328330, 'Freedom_Bracket'] = 'low'
df_2015.loc[df_2015['Freedom'] > 0.328330, 'Freedom_Bracket'] = 'medium'
df_2015.loc[df_2015['Freedom'] >0.549092 , 'Freedom_Bracket'] = 'high'

print(df_2015.head())
print(df_2015.tail())
````

````
#value_counts() is so even since the categorical variable splits were based on the %25, %50, %75
df_2015['Freedom_Bracket'].value_counts()
````

##Lets get Happiness Rankings from Multiple Years of The World Happiness Report

####*How have Country Happiness Ranks have changed over the years?*




###2016 data

````
df_2016 = pd.read_csv ('2016.csv')
df_2016.head()
````

###2017 data

````
df_2017 = pd.read_csv ('2017.csv')
df_2017.head()
````


###Merge 2015 and 2016 Data

https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html


````
#merge 2015 and 2016 on Country, to see Happiness Ranks
df_2015 = df_2015.rename(columns={"Happiness Rank": "2015"})
df_2016 = df_2016.rename(columns={"Happiness Rank": "2016"})
df15 = df_2015[['Country', '2015']]
df16 = df_2016[['Country', '2016']]
result1 = pd.merge(df15, df16, on='Country')
result1
````

###Merge 2015/16 Data with 2017 Data

We need to first change the Happiness.Rank name to Happiness Rank to match the other years data.

####Rename a Column

To rename a column, we use the **rename function**.  Inside the function, we say we want to rename the columns, specify which column name we want to rename, and specify the new name.



```
dataframe.rename(columns={"column_to_be_renamed" : "new_name"})
```



````
df_2017 = df_2017.rename(columns={"Happiness.Rank": "2017"})
df_2017.head(2)
````

````
df17 = df_2017[['Country', '2017']]
result2 = pd.merge(result1, df17, on='Country')
result2.head()
````

###2018 and 2019 data.



````
df_2018 = pd.read_csv (r'2018.csv')
df_2018.head(2)
````

###Change Column Names Before Merging
2018 Happiness Rank is called Overall Rank

````
#rename 'Overall Rank' to '2018' and 'Country or region' to 'Country'
df18 = df_2018.rename(columns={"Overall rank": "2018", "Country or region": "Country"})
df18 = df18[['Country', '2018']]
result3 = pd.merge(result2, df18, on='Country')
result3.head()
````

````
df_2019 = pd.read_csv ('2019.csv')
df_2019.head(2)
````

````
#rename 'Overall Rank' to '2019' and 'Country or region' to 'Country'
df19 = df_2019.rename(columns={"Overall rank": "2019", "Country or region": "Country"})
df19 = df19[['Country', '2019']]
result4 = pd.merge(result3, df19, on='Country')
result4.head()
````

##Transpose Data
###To transpose the dataframe is to flip the columns into rows

**Why?**

* In order to graph this data, ideally the indexing would be the years, rather than the index.  This makes for a great x-axis.  

* Isolate based on Country Columns to get the y-axis Happiness Rank and the x-axis Years.

````
result_transpose = result4.set_index('Country').T
result_transpose
````
